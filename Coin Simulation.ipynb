```python
import random
import numpy as np
import matplotlib.pyplot as plt

# Set the number of coin flips
num_flips = 10

# Initialize counters for heads and tails
heads_count = 0
tails_count = 0

# Simulate coin flips
for _ in range(num_flips):
    # Generate a random number: 0 for tails, 1 for heads
    coin_result = random.randint(0, 1)
    
    # Update the counters based on the coin result
    if coin_result == 0:
        tails_count += 1
    else:
        heads_count += 1

# Calculate the probabilities of heads and tails
prob_heads = heads_count / num_flips
prob_tails = tails_count / num_flips

# Print the simulation results and probabilities
print("Heads count:", heads_count)
print("Tails count:", tails_count)
print("Probability of heads:", prob_heads)
print("Probability of tails:", prob_tails)

# Create a bar plot to visualize the probabilities
outcomes = ("Heads", "Tails")
probabilities = [prob_heads, prob_tails]
y_pos = np.arange(len(outcomes))

# Create bars
plt.bar(y_pos, probabilities, color=["blue", "red"])

# Add labels and title
plt.xlabel("Outcomes")
plt.ylabel("Probability")
plt.title("Probability of Coin Flips")

# Add outcome names on the x-axis
plt.xticks(y_pos, outcomes)

# Show the plot
plt.show()
```
